```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = '0-22_logs'

with open(filepath, 'r') as file:
    content = file.read()

lines = content.split("\n")

step = []
loss = []
for line in lines:
    split = line.split(":")
    if (len(split) == 2 and split[0].isdigit()):
        if (len(step) == 0 or step[-1] != int(split[0])):
            step.append(int(split[0]))
            loss.append(float(split[1]))

step = np.array(step)
loss = np.array(loss)
```

```python
plt.plot(step, loss)
```

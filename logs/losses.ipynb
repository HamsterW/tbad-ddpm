```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
step = []
loss = []

def add_loss(step, loss, filepath):
    with open(filepath, 'r') as file:
        content = file.read()

    lines = content.split("\n")

    for line in lines:
        split = line.split(":")
        if (len(split) == 2 and split[0].isdigit()):
            if (len(step) == 0 or step[-1] != int(split[0])):
                step.append(int(split[0]))
                loss.append(float(split[1]))
    return step, loss

def moving_average(arr, window_size=100):
    averages = []
    for i in range(0, len(arr), window_size):
        # Take a slice of the array, up to 100 elements
        chunk = arr[i:i + window_size]
        # Calculate the average of the chunk
        avg = sum(chunk) / len(chunk)
        # Append the average to the result array
        averages.append(avg)
    return averages

filepath = '0-22_logs'
step, loss = add_loss(step, loss, filepath)

filepath = '23-121_logs'
step, loss = add_loss(step, loss, filepath)

arr = [float(i) for i in loss]  # A sample array of floats (from 1 to 1000)
averages = moving_average(arr, 3000)

print(len(averages))
plt.plot([i for i in range(len(averages))], averages)
```
